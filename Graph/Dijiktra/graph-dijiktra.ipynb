```python
############################################################
# GraphBuilder.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
# YOU WRITE CODE IN THIS FILE
###########################################################

class GraphBuilder:
    def __init__(self, g, f: str): #def __init__(self, g: graph, f: str):
        self._g = g
        # graph object
        self._f = f  # File from which you are building graph
        self._g._g = self._build_graph()

    ############################################################
    # Write code: build_graph
    # Use as many private functions and prvate data you want
    ###########################################################
    def _build_graph(self) -> "None":
        with open(self._f, "r") as file:
            data = file.readlines()
            for aline in data:
                token = aline.split()
                size = len(token)
                if (size < 2) or (size > 3):
                    continue
                tf = token[0]
                tt = token[1]
                if (tf == tt):
                    continue
                w=0
                if size == 3: # weights are given for graph
                    tw = token[2]
                    w = float(tw)
                
                # Ensure both nodes are added to the graph
                f_node = self._g.build_node_and_to_graph(tf)
                t_node = self._g.build_node_and_to_graph(tt)
                
                # self._g._add_edge(tf,tt,w)
                self._g._add_edge(f_node,t_node,w)
```

# DFS USING TIME STAMP

```python
############################################################
# GraphDfs.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
# Depth First serach using TimeStamp
###########################################################
############################################################
# GraphDfs.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
# Depth First serach using TimeStamp
###########################################################
class Timestamp:
    def __init__(self):
        self.inn = 0
        self.out = 0
        
class GraphDfsUsingTimeStamp:
    def __init__(
        self,
        g: "graph",
        graph_name: "string",
        dfs_order: "list of Nodes",
        has_loop: "list of size 1",
        work: "list of size 1",
        dfs_traversal_output_file,
    ):
        ##NOTHING CAN BE CHANGED HERE
        self._g = g
        self._graph_name = graph_name
        self._dfs_order = dfs_order
        self._has_loop = has_loop
        self._has_loop[0] = False
        self._work = work
        self._work[0] = 0
        self._dfs_traversal_output_file = dfs_traversal_output_file

        ##YOU CAN has any number of private varibles and funcions
        self._counter = 0 
        self._a = {}
        self._unvisited_nodes = (set())

        self._dfs()
        self._write_dot()

    ##########################################################
    # Write dot file of DFS traversal
    ##########################################################
    def _write_dot(self):
        output_file = open(self._dfs_traversal_output_file, "w")
        output_file.write("## Jagadeesh Vasudevamurthy ####\n")
        output_file.write("digraph g {\n")
        output_file.write('\tlabel = "[') 
        for current_node in self._dfs_order:
            node_name = current_node.get_name()
            output_file.write(node_name)
            output_file.write(" ")
        if self._has_loop[0]:
            output_file.write('] LOOP"\n')
        else:
            output_file.write('] NOLOOP"\n')
        for current_node in self._dfs_order:
            node_name = current_node.get_name()
            output_file.write("\t")
            output_file.write(node_name)
            output_file.write(" [label= ")
            output_file.write("<")
            output_file.write(node_name)
            output_file.write('<BR /><FONT POINT-SIZE="10">')
            timestamp = self._a[current_node]
            output_file.write(str(timestamp.inn))
            output_file.write("/")
            output_file.write(str(timestamp.out))
            output_file.write("</FONT>>]\n")
    
        graph_type = self._g.get_graph_type()
        if graph_type == GraphType.UNDIRECTED or graph_type == GraphType.WEIGHTED_UNDIRECTED:
            output_file.write("\tedge [dir=none, color = blue]\n")
        else:
            output_file.write("\tedge [color=green]\n")
    
        all_nodes = self._g.list_of_nodes()
        for node in all_nodes:
            node_name_1 = node.get_name()
            fanout_edges = node.fanout_of_a_node()
            for edge in fanout_edges:
                node_name_2 = edge.get_name()
                edge_weight = node.node_get_fanout_node_weight(edge)
                edge_str = "" 
                if graph_type == GraphType.WEIGHTED_UNDIRECTED or graph_type == GraphType.WEIGHTED_DIRECTED:
                    if graph_type == GraphType.WEIGHTED_DIRECTED or (node_name_1 < node_name_2):
                        edge_str = f" {node_name_1} -> {node_name_2} [label = {edge_weight}]"
                elif graph_type == GraphType.DIRECTED or (node_name_1 < node_name_2):
                    edge_str = f" {node_name_1} -> {node_name_2}"
                if edge_str:
                    output_file.write(f"\t{edge_str}\n")
        output_file.write("}\n")
        output_file.close()


    ##########################################################
    # TIME: THETA(V + E)
    # SPACE: O(Largest path) = O(V)
    ##########################################################

    def _update_timestamp(self):
        listnodes = self._g.list_of_nodes()
        for n in listnodes:
            t = Timestamp()
            self._a[n] = t
            self._unvisited_nodes.add(n)

    def _is_visited_node(self, n: "Node"):
        return n not in self._unvisited_nodes

    def _mark_node_visited(self, n: "Node"):
        self._unvisited_nodes.remove(n)

    def _next_unvisited_node_from_set(self):
        return next(iter(self._unvisited_nodes), None)

    def _dfs_r(self, node: "Node", parentnode: "Node"):
        self._work[0] += 1
        if not self._is_visited_node(node):
            self._mark_node_visited(node)
            self._counter += 1
            t = self._a[node]
            t.inn = self._counter
            for fanoutnode in node.fanout_of_a_node():
                if fanoutnode != parentnode:
                    self._dfs_r(fanoutnode, node)
            self._counter += 1
            t.out = self._counter
            self._dfs_order.append(node)
        else:
            t = self._a[node]
            if t.out == 0:
                self._has_loop[0] = True
    def _dfs(self):
        self._update_timestamp()
        while True:
            nextnode = self._next_unvisited_node_from_set()
            if nextnode:
                self._dfs_r(nextnode, nextnode)
            else:
                break
        self._dfs_order.reverse()
```

# Dijkstra Algorithm

```python
############################################################
# GraphDijkstra.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2023
###########################################################

############################################################
# All imports
###########################################################
import heapq

############################################################
# self, gname, start_city, cost, work, dijkstra_traversal_dot_output_file
###########################################################
class GraphDijkstra:
    def __init__(
        self,
        g: "graph",
        graph_name: "string",
        start_city_name: "string",
        cost: "list of Nodes",  # FILL: cost from start city to all other city.If not reachable -1
        work: "list of size 1",  # FILL
        Dijkstra_traversal_output_file: "String",
        show: "bool",
    ):
        ##NOTHING CAN BE CHANGED HERE
        self._g = g
        self._gname = graph_name
        self._start_city_name = start_city_name
        self._cost = cost
        self._work = work
        self._work[0] = 0
        self._Dijkstra_traversal_output_file = Dijkstra_traversal_output_file
        self._show = show

        ##YOU CAN has any number of private varibles and funcions
        self._unvisited_nodes = set()
        self._visited_nodes = {} # visited, updated cost
        self._priority_queue = []
        self._distances = {}

        ###
        self._Dijkstra()
        self._write_dot()
        

    ##########################################################
    # All private function below
    ##########################################################
    def _increment_work(self, c):
        self._work[0] = self._work[0] + c

    def _write_dot(self):
        output = []
        graph_type = self._g.get_graph_type()
        
        # Add a comment about the graph type
        if graph_type == "WEIGHTED_DIRECTED":
            output.append('// This is a weighted directed graph\n')
            output.append('digraph G {\n')
        else:
            output.append('// This is a weighted undirected graph\n')
            output.append('graph G {\n')
        
        # Write nodes
        for node in self._g.list_of_nodes():
            node_name = node.get_name()
            if node in self._distances and self._distances[node] < float('infinity'):
                output.append(f'  "{node_name}" [label="{node_name} ({self._distances[node]})"];\n')
            else:
                output.append(f'  "{node_name}" [label="{node_name}"];\n')
        
        # Write edges
        for node in self._g.list_of_nodes():
            for neighbor in node.fanout_of_a_node():
                weight = node.node_get_fanout_node_weight(neighbor)
                if graph_type == "WEIGHTED_DIRECTED":
                    output.append(f'  "{node.get_name()}" -> "{neighbor.get_name()}" [label="{weight}"];\n')
                else:
                    output.append(f'  "{node.get_name()}" -- "{neighbor.get_name()}" [label="{weight}"];\n')
        
        output.append('}\n')
        
        with open(self._Dijkstra_traversal_output_file, 'w') as f:
            f.writelines(output)


    ##########################################################
    # Dijkstra algorithm
    ##########################################################

    def _get_unvisited_nodes_list(self):
        listnodes = self._g.list_of_nodes()
        self._distances = {node: float('infinity') for node in listnodes}
        for node, costing in self._distances.items():
            if node.get_name().__eq__(self._start_city_name):
                costing = 0.0
                heapq.heappush(self._priority_queue, (costing, node)) # start
                self._distances[node] = costing
            else:
                self._unvisited_nodes.add(node)

    def _Dijkstra(self):
        self._get_unvisited_nodes_list()
        
        while self._priority_queue: 
            current_cost, source_node = heapq.heappop(self._priority_queue)
            self._increment_work(1)

            for dest_node in source_node.fanout_of_a_node(): # 2, 4,
                w = source_node.node_get_fanout_node_weight(dest_node) 
                new_cost = current_cost + w

                if new_cost < self._distances[dest_node]:
                    heapq.heappush(self._priority_queue, (new_cost, dest_node))
                    self._distances[dest_node] = new_cost
                    self._increment_work(1)  

        # end
        for node, cost in self._distances.items():
            if cost == float('inf'):
                cost = -1.0
            self._cost.append(cost)
```

```python
############################################################
# Graph.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
# NOTHING CAN BE CHANGED IN THIS FILE
###########################################################


############################################################
# Edge
# T can be number or name or user key
# Decision taken on other
# Must write __eq__ and __hash__
###########################################################
class Edge:
    def __init__(self, n: "Node", weight: float):
        # All keys below. 
        # Decision is taken based on key. Do not change key after inserted
        self._other = n  # _other is node
        # All values below
        # We can change these values
        self._weight = weight  # _weight is float

    ############################################################
    # All Public routines. YOU SHOULD ONLY CALL THESE ROUTINES
    ###########################################################
    def get_node(self) -> "Node":
        return self._other
    
    def get_name(self) -> "T":
        return self.get_node().get_name()

    def get_weight(self) -> float:
        return self._weight

    def change_weight(self, w: float) -> None:
        self._weight = w

    ############################################################
    # All Private routines. YOU SHOULD NOT CALL THESE ROUTINES
    ###########################################################

    #########  WRITE 5 below starts ################
    def _get_key(self)->'T':
        return self.get_name()

    def __hash__(self)->int:
        return (hash(self._get_key()))

    def __lt__(self, other:'Node')-> bool:
        if self._get_key() < other._get_key():
            return True
        else:
            return False

    def __eq__(self, other)-> bool:
        if (not(self < other) or (other < self)):
            return True
        return False

    def __ne__(self, other)-> bool:
        return (not(self == other))
        
    #########  WRITE 5 below ends ################


############################################################
# Node
# T can be number or name or user key
# Decision taken on T
# Must write __eq__ and __hash__
###########################################################
class Node:
    def __init__(self, T: "T"):
        # node data structures
        # All keys below. 
        # Decision is taken based on key. Do not change key after inserted
        self._item = T  # self._item is our key in hash/set
        # All values below
        # We can change these values
        self._fanins = {}  # dict of fanins of Node. Key is edge other node name, Value is Edge
        self._fanouts = {}  # dict of fanouts of Node. Key is edge other node name , Value is Edge

    ############################################################
    # All Public routines. YOU SHOULD ONLY CALL THESE ROUTINES
    ###########################################################
    def get_name(self) -> "T":
        return self._item

    def add_fan_out(self, e: "Edge") -> None:
        self._fanouts[e] = e

    def add_fan_in(self, e: "Edge") -> None:
        self._fanins[e] = e

    def num_fan_outs(self) -> "int":
        return len(self._fanouts)

    def num_fan_ins(self) -> "int":
        return len(self._fanins)

    def fanout_of_a_node_edges(self) -> "list of fanout edges":
        return list(self._fanouts.values())

    def fanout_of_a_node(self) -> "list of fanout Nodes":
        l = self.fanout_of_a_node_edges()
        n = []
        for edge in l:
            n.append(edge.get_node())
        return n

    def fanin_of_a_node_edges(self) -> "list of fanin edges":
        return list(self._fanins.values())

    def fanin_of_a_node(self) -> "list of fanin Nodes":
        l = self.fanin_of_a_node_edges()
        n = []
        for edge in l:
            n.append(edge.get_node())
        return n

    def node_has_fanout_edge(self, e: "Edge") -> "Edge or None":
        aedge = self._fanouts.get(e)
        if aedge:
            return aedge
        else:
            return None

    def node_has_fanout(self, e: "Edge") -> "Node or None":
        aedge = self.node_has_fanout_edge(e)
        return edge.get_node() if edge else None

    def node_has_fanin_edge(self, e: "Edge") -> "Edge or None":
        return self._fanins.get(e, None)

    def node_has_fanin(self, e: "Edge") -> "Node or None":
        aedge = self.node_has_fanin_edge(e)
        return aedge.get_node() if edge else None

    def node_get_fanout_node_weight(self, other_node: "Node") -> "float":
        nodename = self.get_name()
        aedge = self._fanouts.get(other_node)
        edgename = aedge.get_name()
        assert nodename != edgename
        return aedge.get_weight()

    def node_get_fanin_node_weight(self, other_node: "Node") -> "float":
        # edge = self._fanins.get(other_node.get_name(), None)
        # return edge.get_weight() if edge else -1
        nodename = self.get_name()
        aedge = self._fanins.get(other_node)
        edgename = aedge.get_name()
        assert nodename != edgename
        return aedge.get_weight()
        

    ############################################################
    # All Private routines. YOU SHOULD NOT CALL THESE ROUTINES
    ###########################################################

    #########  WRITE 5 below starts ################
    def _get_key(self) -> "T":
        return self.get_name()

    def __hash__(self) -> "int":
        return hash(self._get_key())

    def __lt__(self, other: 'Node') -> 'bool':
        return self._get_key() < other._get_key()

    def __eq__(self, other: 'Node') -> 'bool':
        # return self._get_key() == other._get_key()
        return (not(self < other)) and (not(other < self))

    def __ne__(self, other: 'Node') -> 'bool':
        return (not(self == other))

    #########  WRITE 5 below ends ################


class Graph:
    ##GRAPH DATA STRUCTURE
    def __init__(self,graphtype: "GraphType"):
        self._numE = 0  # Number of edges
        self._type = graphtype  # Graph type
        self._dict = {}  # Key is Node key, Value is Node


    ############################################################
    # All Public routines. YOU SHOULD ONLY CALL THESE ROUTINES
    ###########################################################
    
    def is_directed_graph(self) -> "bool":
        # return self._type in [GraphType.DIRECTED, GraphType.WEIGHTED_DIRECTED]
        if self._type == GraphType.DIRECTED:
            return True
        if self._type == GraphType.WEIGHTED_DIRECTED:
            return True
        return False

    def is_undirected_graph(self) -> "bool":
        # return self._type in [GraphType.UNDIRECTED, GraphType.WEIGHTED_UNDIRECTED]
        return not (self.is_directed_graph())

    def is_weighted_graph(self) -> "bool":
        # return self._type in [GraphType.WEIGHTED_DIRECTED, GraphType.WEIGHTED_UNDIRECTED]
        if self._type == GraphType.WEIGHTED_DIRECTED:
            return True
        if self._type == GraphType.WEIGHTED_UNDIRECTED:
            return True
        return False

    def get_graph_type(self) -> "GraphType":
        return self._type

    def get_graph_type_as_string(self) -> "string":
        t = self.get_graph_type()
        if t == GraphType.DIRECTED:
            return "DIRECTED GRAPH"
        if t == GraphType.UNDIRECTED:
            return "UNDIRECTED GRAPH"
        if t == GraphType.WEIGHTED_DIRECTED:
            return "WEIGHTED_DIRECTED GRAPH"
        if t == GraphType.WEIGHTED_UNDIRECTED:
            return "WEIGHTED_UNDIRECTED GRAPH"
        return "None"

    def get_numV(self) -> "int":
        return len(self._dict)

    def get_numE(self) -> "int":
        return self._numE  

    def list_of_nodes(self) -> "list of nodes":
        return list(self._dict.values())

    def build_node(self, name: 'T') -> "Node":
        return Node(name)

    def build_node_and_to_graph(self, name: 'T') -> "Node": #add to graph
        n = Node(name)
        self._add_node(n)
        return n

    def has_node(self, node: 'Node') -> "bool":
        return node.get_name() in self._dict

    def get_node(self, node: 'Node') -> 'Node':
        if self.has_node(node):
            n = self._dict[node.get_name()]  # correct key access
            assert n
            return n
        return None

    
    def add_edge(self,fname:'T', tname:'T', w:'float'):
        f = Node(fname)
        t = Node(tname)
        self._add_edge(f,t,w)
        
    ##########################################################
    # NOTHING CAN BE CHANGED IN THIS ROUTINE
    # if this passes means all your interface is right
    #  #Use this routine as a pattern for all other routine
    ##########################################################
    
    def dump(self, name):
        print("------------", name, "------------ ")
        print(self.get_graph_type())
        print("Num Vertices =", self.get_numV())
        print("Num Edges    =", self._numE)
        numedge = 0
        all_nodes = self.list_of_nodes()
        for node in all_nodes:
            n = node.get_name()
            print(n, "Fanouts: ", end="")
            f = node.num_fan_outs()
            if f == 0:
                print("NONE")
            else:
                fanout_nodes = node.fanout_of_a_node()
                j = 0
                for fnode in fanout_nodes:
                    numedge = numedge + 1
                    d = fnode.get_name()
                    if j < f - 1:
                        print(d, ",", sep="", end="")
                    else:
                        print(d)
                    j = j + 1
            print(n, "Fanins : ", end="")
            f = node.num_fan_ins()
            if f == 0:
                print("NONE")
            else:
                fanin_nodes = node.fanin_of_a_node()
                j = 0
                for fnode in fanin_nodes:
                    numedge = numedge + 1
                    d = fnode.get_name()
                    if j < f - 1:
                        print(d, ",", sep="", end="")
                    else:
                        print(d)
                    j = j + 1
        assert numedge/2 == self._numE

    ##########################################################
    # Nothing can be changed
    # TIME: THETA(V + E)
    # SPACE: THETA(V)
    ##########################################################
    def assert_dfs_passed(self, has_loop: "bool", dfs_order: "list of nodes"):
        t = self.get_graph_type()
        if (t == GraphType.UNDIRECTED) or (t == GraphType.WEIGHTED_UNDIRECTED):
            return
        if has_loop == False:
            set_of_visited_nodes = set()
            for n in dfs_order:
                ## Go on fanins of node
                fanins_of_n = n.fanin_of_a_node()
                for nf in fanins_of_n:
                    must_be_there = nf in set_of_visited_nodes  # find in THETA(1)
                    assert must_be_there
                set_of_visited_nodes.add(n)  # add in THETA(1)
            # All nodes must be visited
            assert len(set_of_visited_nodes) == self.get_numV()
            print("DFS ASSERT PASSED")

    ############################################################
    # All Private routines. YOU SHOULD NOT CALL THESE ROUTINES
    ###########################################################
    def _add_node(self, n:'Node')->"Node":
        storedn = self.get_node(n)
        if storedn:
            return storedn
        # self._dict[n] = n # because we have written what is key in node. Hash works.
        self._dict[n.get_name()] = n  # Store using node name as key
        storedn = self.get_node(n)
        assert storedn
        return storedn
        

    def _add_an_edge(self,f:'Node', t:'Node', fanout:'bool', w:'float'):
        # first discover whether its fanin or fanout
        # its with weight or without weight
        f = self._add_node(f)
        t = self._add_node(t)
        if fanout:
            e = f.node_has_fanout_edge(t)
            if e:
                ew = e.get_weight()
                if w < ew:
                    e.change_weight(w)
            else:
                # first time
                e = Edge(t, w)
                self._numE += 1
                f.add_fan_out(e)
        else:
            e = f.node_has_fanin_edge(t)
            if e:
                ew = e.get_weight()
                if w < ew:
                    e.change_weight(w)
            else:
                # first time
                e = Edge(t, w)
                # self._numE += 1
                f.add_fan_in(e)
        
    ##########################################################
    # Nothing can be changed in _add_edge
    # You must write _add_an_edge for this routine to work
    ##########################################################
    def _add_edge(self,f:'Node', t:'Node', w:'float'):
        self._add_an_edge(f,t,True,w) #fanout
        self._add_an_edge(t,f,False,w) #fanin
        # if (self._type == GraphType.UNDIRECTED) or (self._type == GraphType.WEIGHTED_UNDIRECTED):
        if self.is_undirected_graph():
            self._add_an_edge(t,f,True,w) #fanout
            self._add_an_edge(f,t,False,w) #fanin

    ############################################################
    ## All the routines written by students
    ##########################################################
    def build_graph(self, f: "file name"):
        b = GraphBuilder(self, f)  

    def write_dot(self, f):
        b = GraphDot(self, f)

    def show_dot_file(self, filename: "string"):
        with open(filename) as f:
            dot_graph = f.read()
        return dot_graph

    def dfs_using_time_stamp(
        self,
        gname: "string",
        dfs_order: "list of nodes",
        has_loop: "List of size 1 Boolean",
        work: "list of size 1",
        dfs_dot_output_file: "Traversal file name",
    ):

        b = GraphDfsUsingTimeStamp(
            self, gname, dfs_order, has_loop, work, dfs_dot_output_file
        )

    def bfs(
        self,
        gname: "graphname",
        start: "start city name",
        endt: "end city name",
        dist: "list of size 1",
        work: "list of size 1",
        ofile: "bfs_traversal_output_file",
        show: "bool",
    ):
        b = GraphBfs(self, gname, start, endt, dist, work, ofile, show)

    def Dijkstra(
        self,
        gname: "string",
        start_city: "String",
        cost: "list of cost",  # Caller will Fill. cost from start city to all other city.If not reachable -1
        work: "list of size 1",
        dijkstra_traversal_dot_output_file: "Traversal file name",
        show: "bool",
    ):
        b = GraphDijkstra(
            self,
            gname,
            start_city,
            cost,
            work,
            dijkstra_traversal_dot_output_file,
            show,
        )
```
