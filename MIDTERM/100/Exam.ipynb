```python
############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
#  ADD HERE
###########################################################

############################################################
#  ListNode class
#  NOTHING CAN BE CHANGED IN THIS CLASS
###########################################################
class ListNode:
    def __init__(self, val = 0, next= None):
        self.val = val
        self.next = next

    def change_value(self,x:'int')->'None':
        self.val = x

############################################################
#  Slist class
#  WRITE CODE ONLY IN SLIST CLASS
###########################################################
class Slist:
    def __init__(self, n:'Python Int' = None):
        #You must use first and last.
        self._first = None
        self._last = None
        self._positive = True

        if n is not None:
            self._initialize_from_number(n)

    def _initialize_from_number(self, n: int):
        if n < 0:
            self._positive = False
            n = -n
        else:
            self._positive = True
        n_digits = str(n)
        for digit in n_digits:
            new_node = ListNode(int(digit))
            if self._first is not None:
                self._last.next = new_node
                self._last = new_node
            else:
                self._first = new_node
                self._last = new_node

    ######################################################
    # This routine is only for testing. You cannot use it
    # You cannot convert slist to PYTHON INT or any other data structure
    ####################################################
    def Int(self)->'python int':
        s = 0
        sign = self.is_positive()
        n = self._first
        assert(n)
        l = len(self)
        if (l > 1):
            if (n.val == 0):
                print(self)
                print("You cannot have leading zero")
                assert(0)
        if (l == 1 and n.val == 0):
            if (sign == False):
                print(self)
                print("There is no -0")
                assert(0)

        while (n != None):
            v = n.val
            if (v < 0 or v > 9):
                print(self)
                print("YOU MUST STORE SINGLE DIGIT 0 to 9 only with each node")
                assert(0)
            s = (10 * s)  + v
            n = n.next
            if (sign == False):
                s = -s
        return s


    #############################
    # WRITE All public functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    #############################

    #############################
    # YOU MUST WRITE whether the number represented in Slist + or -
    # NOTE THERE IS NO -0. For 0 sign must be positive
    #############################
    def is_positive(self) -> 'Bool':
        return self._positive

    def is_negative(self) -> 'Bool':
        return not self._positive

    def __len__(self) -> 'int':
        n_len = 0
        node = self._first
        while node:
            n_len += 1
            node = node.next
        return n_len

    def __repr__(self):
        if self._first is None:
            return "Empty Slist"

        self._remove_leading_zeros()
        result = "-" if self.is_negative() else ""
        node = self._first
        while node:
            result += str(node.val)
            node = node.next
        return result
    
    
    def _remove_leading_zeros(self):
        while self._first and self._first.val == 0 and self._first.next:
            self._first = self._first.next

        if self._first is None:
            self._last = None
            self._positive = True 

        if self._first is None:
            self._last = None
        else:
            current = self._first
            while current.next is not None:
                current = current.next
            self._last = current

    def __add__(self, other):
        if not isinstance(other, Slist):
            raise TypeError(f"Unsupported operand type(s) for +: 'Slist' and '{type(other).__name__}'")

        if self.is_negative() and other.is_negative():
            result = self._add_positive_numbers(self, other)
            result._positive = False
            return result
        elif self.is_negative():
            return other - (-self)
        elif other.is_negative():
            return self - (-other)

        return self._add_positive_numbers(self, other)

    def __sub__(self, other):
        if not isinstance(other, Slist):
            raise TypeError(f"Unsupported operand type(s) for -: 'Slist' and '{type(other).__name__}'")

        if self.is_negative() and other.is_negative():
            return -((-self) + (-other))
        elif self.is_negative():
            result = self._add_positive_numbers(-self, other)
            result._positive = False
            return result
        elif other.is_negative():
            return self + (-other)

        if self < other:
            result = other._subtract_positive_numbers(other, self)
            result._positive = False
            return result

        return self._subtract_positive_numbers(self, other)
    
    def _reverse_slist(self, lst):
        prev = None
        current = lst._first
        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        reverse_slist = Slist()
        reverse_slist._first = prev
        current = prev
        while current and current.next:
            current = current.next
        reverse_slist._last = current
        return reverse_slist
    

    def _add_positive_numbers(self, list1, list2):
        # Reverse the input lists
        list1 = self._reverse_slist(list1)
        list2 = self._reverse_slist(list2)

        head = ListNode(0)
        current_node = head
        carry = 0

        node1 = list1._first
        node2 = list2._first

        while node1 or node2:
            val1 = node1.val if node1 else 0
            val2 = node2.val if node2 else 0
            total = carry + val1 + val2
            carry = total // 10
            current_node.next = ListNode(total % 10)
            current_node = current_node.next

            if node1:
                node1 = node1.next
            if node2:
                node2 = node2.next

        if carry > 0:
            current_node.next = ListNode(carry)

        result = Slist()
        result._first = head.next

        # Traverse to the end to set the _last pointer
        current_node = result._first
        while current_node and current_node.next:
            current_node = current_node.next
        result._last = current_node

        # Reverse the result list to restore proper order
        return self._reverse_slist(result)


    def _subtract_positive_numbers(self, list1, list2):
        list1 = self._reverse_slist(list1)
        list2 = self._reverse_slist(list2)

        head = ListNode(0)
        node = head
        borrowed = 0

        node1 = list1._first
        node2 = list2._first

        while node1:
            val1 = node1.val
            val2 = node2.val if node2 else 0

            sub = val1 - borrowed - val2
            if sub < 0:
                sub += 10
                borrowed = 1
            else:
                borrowed = 0

            node.next = ListNode(sub)
            node = node.next

            if node1:
                node1 = node1.next
            if node2:
                node2 = node2.next

        result = Slist()
        result._first = head.next
        result._last = node

        result._remove_leading_zeros()
        return self._reverse_slist(result)

    def __neg__(self):
        result = Slist()
        result._first = self._first
        result._last = self._last
        result._positive = not self._positive
        return result

    def __lt__(self, other):
        if not isinstance(other, Slist):
            raise TypeError(f"Unsupported operand type(s) for <: 'Slist' and '{type(other).__name__}'")

        if self.is_negative() and not other.is_negative():
            return True
        if not self.is_negative() and other.is_negative():
            return False

        len1, len2 = len(self), len(other)

        if len1 != len2:
            return len1 < len2

        node1, node2 = self._first, other._first
        while node1 and node2:
            if node1.val != node2.val:
                return node1.val < node2.val
            node1 = node1.next
            node2 = node2.next

        return False
            
```
