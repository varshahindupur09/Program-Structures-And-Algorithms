```python
############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
#  ADD HERE
###########################################################

############################################################
#  ListNode class
#  NOTHING CAN BE CHANGED IN THIS CLASS
###########################################################
class ListNode:
    def __init__(self, val = 0, next= None):
        self.val = val
        self.next = next

    def change_value(self,x:'int')->'None':
        self.val = x

############################################################
#  Slist class
#  WRITE CODE ONLY IN SLIST CLASS
###########################################################
class Slist:
    def __init__(self, n:'Python Int' = None):
        #You must use first and last.
        self._first = None
        self._last = None
        self._is_positive = True
        self._is_positive = True  # True for positive, False for negative

        if n is None:
            return

        if n < 0:
            self._is_positive = False
            n = -n

        digits = [int(d) for d in str(n)]
        for digit in digits:
            new_node = ListNode(digit)
            if self._first is None:
                self._first = new_node
                self._last = new_node
            else:
                self._last.next = new_node
                self._last = new_node

        ## add what ever you want here to handle sign
        # YOU CANNOT STORE n in SLIST DATA STRUCTURE
        #construct Slist using n, digit by digit
        #if n is None, both _first and _last is None
        # if n = 1986 1->9->8->6->None
        # You cannot convert slist to PYTHON INT or any other data structure

        #WRITE YOUR CODE HERE



    ######################################################
    # This routine is only for testing. You cannot use it
    # You cannot convert slist to PYTHON INT or any other data structure
    ####################################################
    def Int(self)->'python int':
        s = 0
        sign = self.is_positive()
        n = self._first
        assert(n)
        l = len(self)
        if (l > 1):
          if (n.val == 0):
            print(self)
            print("You cannot have leading zero")
            assert(0)
        if (l == 1 and n.val == 0):
            if (sign == False):
                print(self)
                print("There is no -0")
                assert(0)

        while (n != None):
          v = n.val
          if (v < 0 or v > 9):
            print(self)
            print("YOU MUST STORE SINGLE DIGIT 0 to 9 only with each node")
            assert(0)
          s = (10 * s)  + v
          n = n.next
        if (sign == False):
          s = -s
        return s


    #############################
    # WRITE All public functions BELOW
    # YOU CAN HAVE ANY NUMBER OF PRIVATE FUNCTIONS YOU WANT
    #############################

    #############################
    # YOU MUST WRITE whether the number represented in Slist + or -
    # NOTE THERE IS NO -0. For 0 sign must be positive
    #############################
    def is_positive(self) -> 'Bool':
        return self._is_positive

    def is_negative(self) -> 'Bool':
        return not self._is_positive

    def __len__(self) -> 'int':
        length = 0
        current_node = self._first
        while current_node:
            length += 1
            current_node = current_node.next
        return length

    def __repr__(self):
        if self._first is None:
            return "Empty Slist"

        result = "-" if self.is_negative() else ""
        current_node = self._first
        while current_node:
            result += str(current_node.val)
            current_node = current_node.next
        return result

    def __add__(self, other):
        if not isinstance(other, Slist):
            raise TypeError("Unsupported operand type(s) for +: 'Slist' and '{}'".format(type(other).__name__))

        if self.is_negative() and other.is_negative():
            result = self._add_positive_lists(self, other)
            result._is_positive = False
            return result
        elif self.is_negative():
            return other - (-self)
        elif other.is_negative():
            return self - (-other)

        return self._add_positive_lists(self, other)

    def _add_positive_lists(self, list1, list2):
        dummy_head = ListNode(0)
        current_node = dummy_head
        carry = 0

        node1 = list1._first
        node2 = list2._first

        while node1 or node2:
            val1 = node1.val if node1 else 0
            val2 = node2.val if node2 else 0
            total = carry + val1 + val2
            carry = total // 10
            current_node.next = ListNode(total % 10)
            current_node = current_node.next

            if node1:
                node1 = node1.next
            if node2:
                node2 = node2.next

        if carry > 0:
            current_node.next = ListNode(carry)

        result = Slist()
        result._first = dummy_head.next
        result._last = current_node if carry == 0 else current_node.next
        return result

    def __neg__(self):
        result = Slist()
        result._first = self._first
        result._last = self._last
        result._is_positive = not self._is_positive
        return result

    def __sub__(self, other):
        if not isinstance(other, Slist):
            raise TypeError("Unsupported operand type(s) for -: 'Slist' and '{}'".format(type(other).__name__))

        if self.is_negative() and other.is_negative():
            return -((-self) + (-other))
        elif self.is_negative():
            result = self._add_positive_lists(-self, other)
            result._is_positive = False
            return result
        elif other.is_negative():
            return self + (-other)

        if self < other:
            result = other._subtract_positive_lists(other, self)
            result._is_positive = False
            return result

        return self._subtract_positive_lists(self, other)

    def _subtract_positive_lists(self, list1, list2):
        dummy_head = ListNode(0)
        current_node = dummy_head
        borrow = 0

        node1 = list1._first
        node2 = list2._first

        while node1:
            val1 = node1.val
            val2 = node2.val if node2 else 0

            sub = (val1 - borrow - val2)
            if sub < 0:
                sub += 10
                borrow = 1
            else:
                borrow = 0

            current_node.next = ListNode(sub)
            current_node = current_node.next

            if node1:
                node1 = node1.next
            if node2:
                node2 = node2.next

        result = Slist()
        result._first = dummy_head.next
        result._last = current_node
        return result

    def __lt__(self, other):
        if not isinstance(other, Slist):
            raise TypeError("Unsupported operand type(s) for <: 'Slist' and '{}'".format(type(other).__name__))

        if self.is_negative() and not other.is_negative():
            return True
        if not self.is_negative() and other.is_negative():
            return False

        len1, len2 = len(self), len(other)

        if len1 != len2:
            return len1 < len2

        node1, node2 = self._first, other._first
        while node1 and node2:
            if node1.val != node2.val:
                return node1.val < node2.val
            node1 = node1.next
            node2 = node2.next

        return False



########################################
# Calling code
########################################
class Exam:
    def __init__(self, a:'Slist', b:'Slist', clist:'Empty Slist', work:'list of size 1 that has int', show:'Bool')->'None':
        if (show):
            print("a =",a)
            print("b =",b)
            
            # Increment the work counter
            work[0] += 1

            # Perform the addition
            result = a + b

            # Assign the result to clist
            clist._first = result._first
            clist._last = result._last
            clist._is_positive = result._is_positive

            if show:
                print("a + b =", clist)
```
