```python
############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
# Exam.py
# Author: Jagadeesh Vasudevamurthy 2024
###########################################################

class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

    def change_value(self, x: 'int') -> 'None':
        self.val = x

class Slist:
    def __init__(self, n: 'Python Int' = None):
        self._first = None
        self._last = None
        self._is_positive = True

        if n is None:
            return

        if n < 0:
            self._is_positive = False
            n = -n

        digits = [int(d) for d in str(n)]
        for digit in digits:
            new_node = ListNode(digit)
            if self._first is None:
                self._first = new_node
                self._last = new_node
            else:
                self._last.next = new_node
                self._last = new_node
                
    ######################################################
    # This routine is only for testing. You cannot use it
    # You cannot convert slist to PYTHON INT or any other data structure
    ####################################################
    def Int(self)->'python int':
        s = 0
        sign = self.is_positive()
        n = self._first 
        assert(n)
        l = len(self)
        if (l > 1):
          if (n.val == 0):
            print(self)
            print("You cannot have leading zero")
            assert(0)
        if (l == 1 and n.val == 0):
            if (sign == False):
                print(self)
                print("There is no -0")
                assert(0)
        
        while (n != None):
          v = n.val 
          if (v < 0 or v > 9):
            print(self)
            print("YOU MUST STORE SINGLE DIGIT 0 to 9 only with each node")
            assert(0)
          s = (10 * s)  + v
          n = n.next
        if (sign == False):
          s = -s
        return s
    
    def is_positive(self) -> 'Bool':
        return self._is_positive

    def is_negative(self) -> 'Bool':
        return not self._is_positive

    def __len__(self) -> 'int':
        length = 0
        current_node = self._first
        while current_node:
            length += 1
            current_node = current_node.next
        return length

    def __repr__(self):
        if self._first is None:
            return "Empty Slist"

        self._remove_leading_zeros()
        result = "-" if self.is_negative() else ""
        current_node = self._first
        while current_node:
            result += str(current_node.val)
            current_node = current_node.next
        return result

    def __add__(self, other):
        if not isinstance(other, Slist):
            raise TypeError(f"Unsupported operand type(s) for +: 'Slist' and '{type(other).__name__}'")

        if self.is_negative() and other.is_negative():
            result = self._add_positive_lists(self, other)
            result._is_positive = False
            return result
        elif self.is_negative():
            return other - (-self)
        elif other.is_negative():
            return self - (-other)

        return self._add_positive_lists(self, other)

    def __sub__(self, other):
        if not isinstance(other, Slist):
            raise TypeError(f"Unsupported operand type(s) for -: 'Slist' and '{type(other).__name__}'")

        if self.is_negative() and other.is_negative():
            return -((-self) + (-other))
        elif self.is_negative():
            result = self._add_positive_lists(-self, other)
            result._is_positive = False
            return result
        elif other.is_negative():
            return self + (-other)

        if self < other:
            result = other._subtract_positive_lists(other, self)
            result._is_positive = False
            return result

        return self._subtract_positive_lists(self, other)

    def _add_positive_lists(self, list1, list2):
        list1 = self._reverse_list(list1)
        list2 = self._reverse_list(list2)
        
        dummy_head = ListNode(0)
        current_node = dummy_head
        carry = 0

        node1 = list1._first
        node2 = list2._first

        while node1 or node2:
            val1 = node1.val if node1 else 0
            val2 = node2.val if node2 else 0
            total = carry + val1 + val2
            carry = total // 10
            current_node.next = ListNode(total % 10)
            current_node = current_node.next

            if node1:
                node1 = node1.next
            if node2:
                node2 = node2.next

        if carry > 0:
            current_node.next = ListNode(carry)

        result = Slist()
        result._first = dummy_head.next
        result._last = current_node if carry == 0 else current_node.next

        return self._reverse_list(result)

    def _subtract_positive_lists(self, list1, list2):
        list1 = self._reverse_list(list1)
        list2 = self._reverse_list(list2)
        
        dummy_head = ListNode(0)
        current_node = dummy_head
        borrow = 0

        node1 = list1._first
        node2 = list2._first

        while node1:
            val1 = node1.val
            val2 = node2.val if node2 else 0

            sub = (val1 - borrow - val2)
            if sub < 0:
                sub += 10
                borrow = 1
            else:
                borrow = 0

            current_node.next = ListNode(sub)
            current_node = current_node.next

            if node1:
                node1 = node1.next
            if node2:
                node2 = node2.next

        result = Slist()
        result._first = dummy_head.next
        result._last = current_node

        result._remove_leading_zeros()
        return self._reverse_list(result)

    def __neg__(self):
        result = Slist()
        result._first = self._first
        result._last = self._last
        result._is_positive = not self._is_positive
        return result

    def __lt__(self, other):
        if not isinstance(other, Slist):
            raise TypeError(f"Unsupported operand type(s) for <: 'Slist' and '{type(other).__name__}'")

        if self.is_negative() and not other.is_negative():
            return True
        if not self.is_negative() and other.is_negative():
            return False

        len1, len2 = len(self), len(other)

        if len1 != len2:
            return len1 < len2

        node1, node2 = self._first, other._first
        while node1 and node2:
            if node1.val != node2.val:
                return node1.val < node2.val
            node1 = node1.next
            node2 = node2.next

        return False

    def _remove_leading_zeros(self):
        while self._first and self._first.val == 0 and self._first.next:
            self._first = self._first.next

        if self._first is None:
            self._last = None
            self._is_positive = True  # 0 should be positive

        if self._first is None:
            self._last = None
        else:
            current = self._first
            while current.next is not None:
                current = current.next
            self._last = current

    def _reverse_list(self, lst):
        prev = None
        current = lst._first
        while current is not None:
            next_node = current.next
            current.next = prev
            prev = current
            current = next_node
        lst._first = prev
        return lst

class Exam:
    def __init__(self, a: 'Slist', b: 'Slist', clist: 'Empty Slist', work: 'list of size 1 that has int', show: 'Bool') -> 'None':
        if show:
            print("a =", a)
            print("b =", b)

        work[0] += 1
        result = a + b

        clist._first = result._first
        clist._last = result._last
        clist._is_positive = result._is_positive

        if show:
            print("a + b =", clist)
############################################################
# ExamTest.py
# Author: Jagadeesh Vasudevamurthy
# Copyright: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
#     NOTHING CAN BE CHANGED IN THIS FILE
###########################################################

############################################################
# All imports
###########################################################
# from Exam import *

############################################################
# ExamTest.py
# Author: Jagadeesh Vasudevamurthy 2024
###########################################################

############################################################
#     NOTHING CAN BE CHANGED IN THIS FILE
###########################################################

############################################################
# All imports
###########################################################
# from Exam import *

class ExamTest:
    def __init__(self):
        self._no = 0
        self._show = False
        self._testBench()

    def _testBench(self):
        self.advanced_tests()
        print("All basic tests passed")
        print("You got 50 marks")
        print("real test bench given after exam")

    def _test1(self, a: 'int', b: 'int') -> 'None':
        if self._show:
            print("----------- Test case ", self._no, "----------------")
        self._no += 1
        c = a + b
        if self._show:
            print("a = ", a, "b = ", b, "a+b =", c)
            print("---- Let us see your work starts ---")

        alist = Slist(a)
        blist = Slist(b)
        clist = Slist()
        work = [0]  # work done by algorithm
        s = Exam(alist, blist, clist, work, self._show)
        if self._show:
            print("ans = ", clist, "Len = ", len(clist), "work =", work[0])
            if clist.is_positive():
                print("positive")
            if clist.is_negative():
                print("negative")

        if self._show:
            print("ans = ", clist, "work =", work[0])
        x = clist.Int()
        if c != x:
            print("Your Answer is", x, " But expected answer is", c)
            assert(0)
        if work[0] == 0:
            print("How did you add without work")
            assert(0)

   

    def advanced_tests(self):
        self._show = True
        test_cases = [
            # Edge cases
            (0, 0),
            (1, 0),
            (0, -1),
            (-1, 1),

            # Large numbers
            (999999, 1),
            (1000000, -1),
            (-999999, -1),

            # Numbers with different number of digits
            (12345, 67),
            (9876, -1234),

            # Numbers causing multiple carries
            (9999, 1),
            (9999, 9999),

            # Numbers causing multiple borrows
            (10000, -1),
            (10000, -9999),

            # Very large numbers
            (1234567890, 9876543210),
            (-9876543210, 1234567890),
        ]

        for a, b in test_cases:
            self._test1(a, b)

############################################################
# main
# YOU CANNOT CHANGE ANYTHING BELOW
###########################################################
def main():
    print("Testing ExamTest.py Starts")
    s = ExamTest()
    print("All tests passed. You are a genius")
    print("If you liked this problem send me a box of chocolates")
    print("FILL Complexity of my algorithm. Will lose 10 marks if you don't fill:---------------------------------")
    print("Testing ExamTest.py ENDS")

############################################################
# main
############################################################
if (__name__ == '__main__'):
    main()
```
