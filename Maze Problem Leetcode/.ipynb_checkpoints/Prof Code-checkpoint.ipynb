```python
def two_sum(nums, target):
    seen = {}  # Use a dictionary to store seen numbers and their indices
    result = []  # List to store pairs of indices

    for index, num in enumerate(nums):
        complement = target - num
        if complement in seen:
            for idx in seen[complement]:
                result.append([idx, index])
        seen[num] = [index]
    
    return result

# Example usage:
nums = [2, 7, 11, 5, 15, 4, 1, 8, 16, 18, 0, 20, 100, 9]
idx_list = {}
for index, num in enumerate(nums):
    idx_list[num]=index
print(idx_list)
target = 9
print(two_sum(nums, target))  # Output: [[0, 1], [5, 4]]
```

```python
class TwoSum:
    def __init__(self):
        self.seen = {}  # Dictionary to store seen numbers and their indices

    def two_sum(self, nums, target):
        result = []  # List to store pairs of indices

        for index, num in enumerate(nums):
            complement = target - num
            if complement in self.seen: # compl = seen data
                for idx in self.seen[complement]:
                    result.append([idx, index])
            if num in self.seen:
                self.seen[num].append(index)
            else:
                self.seen[num] = [index]
        
        return result

# Example usage:
nums = [2, 7, 11, 5, 15, 4, 1, 8, 16, 18, 0, 20, 100, 9]
two_sum_instance = TwoSum()
target = 9
print(two_sum_instance.two_sum(nums, target))  # Output: [[0, 1], [5, 4]]
```
