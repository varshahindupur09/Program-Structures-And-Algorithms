Max Value of 9 can obtained from items {3,1} that has values {4+5=9}

```python
class knapsack:
    def __init__(self, s, w, v, ans, work):
        self._w = w
        self._v = v
        self._s = s
        self._work = work
        self._weights_selected = []
        self._values_of_weights_selected = []

        self._knapsack_function(ans)

    def _knapsack_function(self, ans): 
        n = len(self._w)  # no. of items
        V = [[0 for _ in range(self._s + 1)] for _ in range(n + 1)]
        K = [[0 for _ in range(self._s + 1)] for _ in range(n + 1)]
    
        for item in range(1, n + 1):
            for currweight in range(1, self._s + 1):
                self._increment_work()  # Incrementing work counter for each iteration
                if currweight >= self._w[item - 1]: 
                    curr = self._v[item - 1]
                    remain = currweight - self._w[item - 1]
                    previous = V[item - 1][currweight]
                    steal = self._v[item - 1] + V[item - 1][remain]
                    if steal > previous:
                        K[item][currweight] = 1
                    max_value = max(steal, previous)
                    V[item][currweight] = max_value

        # Set the maximum value in ans[0]
        ans[0] = V[n][self._s]

        # calling retracing function
        self._retrace_get_max(V, ans)

        # print the output
        i, w_str, v_str = "", "", ""
        for it in range(len(self._w)):
            i += str(it + 1) + " "
            w_str += str(self._w[it]) + " "
            v_str += str(self._v[it]) + " "
        print("i =   ", i)
        print("w =   ", w_str)
        print("v =   ", v_str)
        
        print("------------ V matrix ------------ ")
        v_matrix = ""
        for row in V:
            v_matrix += " ".join(str(x) for x in row) + "\n" 
        print(v_matrix)
        print("------------ K matrix ------------ ")
        k_matrix = ""
        for row in K:
            k_matrix += " ".join(str(x) for x in row) + "\n"
        print(k_matrix)

        # printing final output
        self._final_answer_print(ans)
        print()

    def _retrace_get_max(self, V, ans):
        n = len(self._w)
        w1 = [0] + self._w 
        v1 = [0] + self._v 
        list_selected_values = {}  
      
        item = n
        cw = self._s
        
        while item > 0 and cw > 0:
            if V[item][cw] != V[item - 1][cw]:
                list_selected_values[w1[item], item] = v1[item]
                cw -= w1[item]
            item -= 1
    
        max_val = sum(list_selected_values[e] for e in list_selected_values)
        if max_val == ans[0]:
            pass
        else: 
            ans[0] = 0

        self._weights_selected = [q[0] for q in list_selected_values.keys()]
        self._values_of_weights_selected = list(list_selected_values.values())


    def _final_answer_print(self, ans):
        final_answer_displayed = "+".join(map(str, self._values_of_weights_selected))
        print(f"Max Value of {ans[0]} can be obtained from items {self._weights_selected} that have values {final_answer_displayed} = {ans[0]}")
        print(f"Work Done is: {self._work}")

    def _increment_work(self)->'None':
        self._work[0] += 1

```
